```python
df = pd.read_json('/content/yelp_academic_dataset_business_cleaned.json', lines=True, orient='records')
df.head()
```

```python
def make_text(row):
    text = "{} is located at {}. It has a rating of {} with {} reviews. The cusinine served by the restaurant is {}.".format(row['name'], row['address'],row['stars'], row['review_count'], row['categories'])
    hours = row['hours']
    if hours is not None:
        timings = ', '.join([str(x)+' - '+str(y) for x,y in hours.items() ])
        text += ' The restaurant timings are {}.'.format(timings)

    d = row['attributes']
    if d is not None:
        services = ', '.join([str(x) for x in d if d[x]=='True'])
        text += ' The restaurant offers these services - {}'.format(services)
    return text


df['text'] = df.apply(lambda x: make_text(x),axis=1)
```

```python
docs = list(df['text'][0:5].values)
```

```python
print(docs)
```

```python
documents = [Document(text=t) for t in docs]
```

```python
from llama_index.prompts.prompts import SimpleInputPrompt

system_prompt = "You are a restaurant Q&A assistant. Your goal is to answer questions as accurately as possible based on the instructions and context provided."

# This will wrap the default prompts that are internal to llama-index
prompt = """[INST] <<SYS>>
You are a restaurant Q&A assistant. Your goal is to answer questions as accurately as possible based on the instructions and context provided.
If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information.
<</SYS>>

{prompt} [/INST] {model_reply} [INST] {prompt} [/INST]"""
# query_wrapper_prompt = SimpleInputPrompt(prompt)
query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"bnb_4bit_compute_dtype":torch.bfloat16, "load_in_4bit":True}
)


# tokenizer = AutoTokenizer.from_pretrained('sidharthaM/llama_yelp_data', trust_remote_code=True)
# tokenizer.pad_token = tokenizer.eos_token
# tokenizer.padding_side = "right"

# pipe = pipeline(task="text-generation", model='sidharthaM/llama_yelp_data', tokenizer=tokenizer, max_length=200, device='cuda')
```

```python
embed_model = LangchainEmbedding(
  HuggingFaceEmbeddings()
)
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
index = VectorStoreIndex.from_documents(documents, service_context=service_context)
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What cusine Perkiomen Valley Brewery serves?")
print(response.response)
```

```python
from llama_index.retrievers import BaseRetriever, VectorIndexRetriever, KeywordTableSimpleRetriever
from llama_index.response_synthesizers import get_response_synthesizer
from llama_index.indices.postprocessor import SimilarityPostprocessor

from llama_index.query_engine import RetrieverQueryEngine
```

```python
index = GPTVectorStoreIndex.from_documents(documents, service_context=service_context)
index.storage_context.persist(persist_dir="./index")

# configure retriever
retriever = VectorIndexRetriever(
    service_context=service_context,
    index=index,
    similarity_top_k=1,
    alpha = 0.5
)
```

```python
# configure response synthesizer
response_synthesizer = get_response_synthesizer(
        service_context=service_context,
    )
# retriever = index.as_retriever(retriever_mode='default')
```

```python
# assemble query engine
query_engine = RetrieverQueryEngine(
        retriever=retriever,
        response_synthesizer=response_synthesizer,
    )
```

```python
response = query_engine.query("Suggest me some restaurants.")
print(response)
```
